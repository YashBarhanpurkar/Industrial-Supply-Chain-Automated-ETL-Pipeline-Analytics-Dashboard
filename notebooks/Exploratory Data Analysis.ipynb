```python
region_cleaning_map = {
    "South of  USA": "South US",  # Fixes the double-space typo natively
    "West of USA": "West US",
    "East of USA": "East US",
    "US Center": "Central US"
}

df['Order Region'] = df['Order Region'].str.strip()
df['Order Region'] = df['Order Region'].replace(region_cleaning_map)
```

```python
print(df['Order Region'].value_counts())
```

```python
country_translation = {
    'EE. UU.': 'United States',
    'Estados Unidos': 'United States',
    'Francia': 'France',
    'Alemania': 'Germany',
    'México': 'Mexico',
    'Australia': 'Australia'
    # Add other explicit translations as needed based on your unique values
}

# Apply mapping with a fallback to the original name if it's not in the dictionary
df['Customer Country'] = df['Customer Country'].map(country_translation).fillna(df['Customer Country'])
df['Order Country'] = df['Order Country'].map(country_translation).fillna(df['Order Country'])
```
